```python
df1 = pd.read_csv(input(''), sep = '|')
df1
# In order to separate the customer recordsinto columns I used the separator parameter to display the values without the pipe symbol 
# Each record is then separated where the pipe symbol is located and separated into their perspective columns
# Now, if I do it this way, I can make it easier to preview the data
# call on the variable to display
```

```python
df1.info()
# checking the structure after having separated the customer records into columns using the info attribute 
# This will show the summary infromation about df1 to help me identify any irregularities in the columns
# Here I can tell I have 521 records
# I can tell have a total of 7 columns
# The Dtype for most of this data is string except for the customer id which an integer of 64 bytes
# The amount of non-null values or any value that is not none is 521 entries compared to 520 non null values in the message opt out column where there could be a missing value and 520 phone non null values in phone which means there could be a missing phone number int hat column
# I will then need to check the contents
```

```python
df1.head(51)
# looking at the first 50 records with .head() attribute. With this, I'm able to see that the name values have stray characters
# Will need to take out the caret symbol(^) also the (&!) symbols
# Format the phone numbers correctly 
# Decide what to do with the missing valyes in the message opt out column
```

```python
df1["name"] = df1["name"].str.replace("^","")
df1
# I replaced the caret symbol (^) first by pulling the name column in brackets, this will specify the column I want to make changes to 
# I then applied the .str method to make sure the output we get in return is also a string, along with the. replace method to replace the caret symbol with nothing since its at the very end of the name string.
# I then went ahead and assigned it to the same dataframe and same column to save the changes
# Finally I called on it to view the output
```

```python
df1["name"] = df1["name"].str.replace("&!",".")
df1
# I then replaced the (&) and (!) symbol together since I seen that both of them are together for the first 50 records. 
# This time I replaced it with a period since I noticed they were in between a first intial followed by a last name
# I again pulled up the name column in brackets, this will specify the column I want to make changes to
# I then applied the .str method again to make sure the output we get in return is also a string, along with the. replace method
# I then went ahead and assigned it to the same dataframe and same column to save the changes
# Finally, I called on it to view the output
```
